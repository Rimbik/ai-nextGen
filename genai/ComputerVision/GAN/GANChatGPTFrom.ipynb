```python
import torch
import torch.nn as nn
import torchvision.transforms as transforms
from torchvision.utils import save_image
from PIL import Image
import requests
from io import BytesIO

# -----------------------------
# 1. Define Generator (DCGAN)
# -----------------------------
class Generator(nn.Module):
    def __init__(self, latent_dim=100, img_channels=3, feature_g=64):
        super(Generator, self).__init__()
        self.net = nn.Sequential(
            nn.ConvTranspose2d(latent_dim, feature_g * 8, 4, 1, 0, bias=False),
            nn.BatchNorm2d(feature_g * 8),
            nn.ReLU(True),
            nn.ConvTranspose2d(feature_g * 8, feature_g * 4, 4, 2, 1, bias=False),
            nn.BatchNorm2d(feature_g * 4),
            nn.ReLU(True),
            nn.ConvTranspose2d(feature_g * 4, feature_g * 2, 4, 2, 1, bias=False),
            nn.BatchNorm2d(feature_g * 2),
            nn.ReLU(True),
            nn.ConvTranspose2d(feature_g * 2, img_channels, 4, 2, 1, bias=False),
            nn.Tanh()
        )

    def forward(self, z):
        return self.net(z)

# -----------------------------
# 2. Load Pretrained Weights from URL
# -----------------------------
latent_dim = 100
device = "cuda" if torch.cuda.is_available() else "cpu"
generator = Generator(latent_dim=latent_dim).to(device)

# Example URL for pretrained weights (replace with actual URL)
url = "https://github.com/pytorch/examples/raw/main/dcgan/weights/netG_epoch_199.pth"
response = requests.get(url)

#save the pth
# with open("netG_epoch_199.pth", "wb") as f:
#     f.write(response.content)


state_dict = torch.load(BytesIO(response.content), map_location=device, weights_only=False)
generator.load_state_dict(state_dict)
generator.eval()

# # -----------------------------
# # 3. Prepare Source Image
# # -----------------------------
# transform = transforms.Compose([
#     transforms.Resize(64),
#     transforms.CenterCrop(64),
#     transforms.ToTensor(),
#     transforms.Normalize([0.5]*3, [0.5]*3)
# ])

# source_img = Image.open("sourceHorse.jpg").convert("RGB")
# source_tensor = transform(source_img).unsqueeze(0).to(device)

# # -----------------------------
# # 4. Generate New Image
# # -----------------------------
# z = torch.randn(1, latent_dim, 1, 1, device=device)
# with torch.no_grad():
#     fake_img = generator(z)

# save_image(fake_img, "generated_image.png", normalize=True)
# print("New image generated and saved as generated_image.png")
```
